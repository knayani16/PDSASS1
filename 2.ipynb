```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

dataset_path = '/content/drive/MyDrive/raw_student_data.csv'
student_data = pd.read_csv(dataset_path)

folders = ['data_raw', 'data_cleaned', 'visualizations']

for folder in folders:
    if not os.path.exists(folder):
        os.makedirs(folder)

cleaned_data = student_data.copy()

cleaned_data.to_csv('data_cleaned/cleaned_data.csv', index=False)
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='gender', data=cleaned_data)
plt.title('Gender Distribution')
plt.savefig('visualizations/gender_distribution.png')
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(x='gender', y='math score', data=cleaned_data)
plt.title('Math Scores by Gender')
plt.savefig('visualizations/math_scores_gender.png')
plt.show()

plt.figure(figsize=(12, 6))
sns.countplot(x='parental level of education', data=cleaned_data, order=cleaned_data['parental level of education'].value_counts().index)
plt.title('Distribution of Parental Education Levels')
plt.xticks(rotation=45, ha='right')
plt.savefig('visualizations/parental_education_distribution.png')
plt.show()

plt.figure(figsize=(10, 8))
sns.scatterplot(x='reading score', y='writing score', hue='gender', data=cleaned_data)
plt.title('Reading vs. Writing Scores')
plt.savefig('visualizations/reading_writing_scores.png')
plt.show()

numerical_vars = ['math score', 'reading score', 'writing score']

plt.figure(figsize=(12, 8))
sns.pairplot(cleaned_data[numerical_vars])
plt.suptitle('Pair Plot for Numerical Variables', y=1.02)
plt.savefig('visualizations/pair_plot.png')
plt.show()
```
